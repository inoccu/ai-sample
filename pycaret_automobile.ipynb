```python
# 使用するデータセット（CSVファイル）の読み込み
# CSVファイルは先頭行にタイトルをセットします
# 数値に桁区切りのカンマや単位を含めないようにします
import pandas as pd

df = pd.read_csv('autos_titled_clean.csv')
df.head()
```

```python
# データを訓練用とテスト用に分割します
data = df.sample(frac=0.8, random_state=111)
data_unseen = df.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
# Pycaretで再帰（予測）用アルゴリズムを読み込みます
from pycaret.regression import *

# targetの値として、予測したい項目名を指定します
reg01 = setup(data=data, target='price')
```

```python
# 様々なアルゴリズムでモデルを作成し、評価します
best_model = compare_models()
```

```python
# 最も評価の良いアルゴリズムを選び、さらにトレーニングを行います
model = create_model('ada')
```

```python
# ハイパーパラメータのチューニングを行います
tuned = tune_model(model)
```

```python
# チューニング済みのモデルを評価します
evaluate_model(tuned)
```

```python
# チューニング済みのモデルをファイナライズします
final = finalize_model(tuned)
print(final)
```

```python
# テストデータを用いて予測を行います
unseen_predictions = predict_model(final, data=data_unseen)
unseen_predictions.head()
```

```python
# テストデータを用いた予測結果で評価を行います
from pycaret.utils import check_metric
check_metric(unseen_predictions.price, unseen_predictions.Label, 'R2')
```
